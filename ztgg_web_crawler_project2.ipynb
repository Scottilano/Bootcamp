 #####Firstly we are going to import requests library. Requests allows you to send organic, grass-fed HTTP/1.1 requests, without the need for manual labor.

```python
import requests
website_raw = requests.get('https://en.wikipedia.org/wiki/List_of_Asian_countries_by_area').text
```

```python
print(website_raw)
print(len(website_raw))
```

### We begin by reading the source code for a given web page and creating a BeautifulSoup (soup)object with the BeautifulSoup function. Beautiful Soup is a Python package for parsing HTML and XML documents. It creates a parse tree for parsed pages that can be used to extract data from HTML, which is useful for web scraping. Prettify() function in BeautifulSoup will enable us to view how the tags are nested in the document.

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(website_raw, 'html.parser')
print(soup.prettify())
```

### first task is to find class ‘wikitable sortable’ in the HTML script.

```python
wikitable = soup.find('table',{'class':'wikitable sortable'})
wikitable
```

### Now to extract all the links within <code><a\></code> , we will use find_all().

```python
links = wikitable.findAll('a')
links
```

### next task, from the links, we have to extract the title which is the name of countries.

```python
countries = []
for link in links:
    countries.append(link.get('title'))
    
print(countries)
```

### Convert the list countries into Pandas DataFrame to work in python.

```python
import pandas as pd
df = pd.DataFrame()
df['Country'] = countries
df
```

```python
url = "https://en.wikipedia.org/wiki/List_of_Presidents_of_the_United_States"
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')
tb = soup.find('table', class_='wikitable')
```

```python
tb
```

```python
tb.find_all('b')
```

```python
for link in tb.find_all('b'):
    name = link.find('a')
    print(name.get_text('title'))
```
